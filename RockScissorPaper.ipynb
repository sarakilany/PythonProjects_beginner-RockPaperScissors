```python
'''
Rock Paper Scissors
'''

import random

def rockPaperScissors():
    print("Insert 100 to exit or")
    computerChoice = random.randint(1,3)
    #1:scissor
    #2: paper
    #3:rock
    
    
    while True:
        userChoice = int(input("Insert number from 1 to 3 to choose: 1 is scissors, 2 is paper, and 3 is rock"))
        
        
        if userChoice ==1:
            if computerChoice ==2:
                print("Computer choose paper, you win")
            elif computerChoice ==1:
                print("Computer choose scissors, This is a tie")
            else:
                print("Computer choose rock, you lose")
        elif userChoice ==2:
            if computerChoice ==1:
                print("Computer choose scissors, you win")
            elif computerChoice ==2:
                print("Computer choose paper, This is a tie")
            else:
                print("Computer choose rock, you lose")
        elif userChoice ==3:
            if computerChoice ==1:
                print("Computer choose scissors, you win")
            elif computerChoice ==3:
                print("Computer choose rock, This is a tie")
            else:
                print("Computer choose paper, you lose")
        elif userChoice ==100:
            return None;

        else:
            print("Error, number out of range, try again or insert 100 to exit")
```
